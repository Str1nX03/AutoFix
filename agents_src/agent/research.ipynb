```python
import sys
import os
import uuid

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..', '..')))

from agents_src.agent.support_agent import SupportAgent

print("Initializing Support Agent...")
agent = SupportAgent()

session_id = str(uuid.uuid4())

print("Support Agent Ready! Type 'exit' or 'quit' to stop.")
print("-" * 50)

while True:
    user_input = input("You: ")
    
    if user_input.lower() in ['exit', 'quit']:
        print("\nEnding conversation. Goodbye!")
        break
        
    if not user_input.strip():
        continue
        
    try:
        result = await agent.arun(query=user_input, session_id=session_id)
        
        response = result.get("support_response", "Sorry, I didn't get a response.")
        print(f"\nYou: {user_input}\n")
        print(f"\nAgent: {response}\n")
        print("====" * 20)
        
    except Exception as e:
        print(f"\nError occurred: {e}\n")
```
